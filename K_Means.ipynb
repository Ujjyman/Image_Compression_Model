Importing Libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

Loading dataset

```python
data = pd.read_csv("Aggregation.txt", delimiter="\t",header=None)
data.columns=["X-Coord","Y-Coord","Cluster"]
```

```python
data.shape
```

```python
data.drop("Cluster", axis=1, inplace=True)
```

```python
data.head()
```

Scaling the data as K Means is a Distance Based Algorithm


```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

Scatter Plot Of Dataset

```python
plt.scatter(data_scaled[:, 0], data_scaled[:, 1],c="green")
plt.xlabel("X-Coord")
plt.ylabel("Y-Coord")
plt.title("Scatter Plot of X and Y Coordinates")
```

Applying Elbow Method To Find Number Of Clusters

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(data_scaled)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

Number Of Clusters =6

```python
kmeans = KMeans(n_clusters=6, init='k-means++', random_state=42)
kmeans.fit(data_scaled)
```

Cluster Number of each datapoint

```python
labels = kmeans.labels_
print(labels)
```

```python
data['Cluster'] = labels
```

```python
data.head()
```

Location Of Centroids

```python
centers = kmeans.cluster_centers_
print("Cluster Centers:", centers)
```

Clustered DataPoints


```python

plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=labels,cmap='plasma')
plt.scatter(centers[:, 0], centers[:, 1], c='green', marker='o', s=150)
plt.xlabel("X-Coord")
plt.ylabel("Y-Coord")
plt.title("K-Means with 6 Clusters")
```

```python
silhouette_avg = silhouette_score(data_scaled, kmeans.labels_)

print("Silhouette Index:", silhouette_avg)
```

# SPIRAL **DATASET**

```python
data_spiral = pd.read_csv("spiral.txt", delimiter="\t",header=None)
data_spiral.columns=["X-Coords","Y-Coords","Clusters"]
```

```python
data_spiral.head()
```

```python
data_spiral_scaled=scaler.fit_transform(data_spiral)
```

```python
plt.scatter(data_spiral_scaled[:, 0], data_spiral_scaled[:, 1])
plt.xlabel("X-Coord")
plt.ylabel("Y-Coord")
plt.title("Spiral dataset")
```

```python
kmeans_spiral = KMeans(n_clusters=3, init='k-means++', random_state=42)
kmeans_spiral.fit(data_spiral_scaled)
```

```python
labels_spiral = kmeans_spiral.labels_
print(labels_spiral)
```

```python
centers_spiral = kmeans_spiral.cluster_centers_
print("Cluster Centers:", centers_spiral)
```

```python
plt.scatter(data_spiral_scaled[:, 0], data_spiral_scaled[:, 1], c=labels_spiral,cmap='inferno')
plt.scatter(centers_spiral[:, 0], centers_spiral[:, 1], c='green', marker='o', s=150)
```

```python
silhouette_spiral_avg = silhouette_score(data_spiral_scaled, labels_spiral)

print("Silhouette Index:", silhouette_spiral_avg)
```
