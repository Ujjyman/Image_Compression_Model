# AGGREGATION DATASET

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
data=pd.read_csv("Aggregation.txt",delimiter="\t",header=None)
data.columns=["X-Coordinate","Y-Coordinate","Cluster"]
data.drop("Cluster",axis=1,inplace=True)
```

```python
data.head()
```

```python
scaler=StandardScaler()
data_scaled=scaler.fit_transform(data)
```

```python
plt.scatter(data_scaled[:, 0], data_scaled[:, 1])
plt.xlabel("X-Coordinate")
plt.ylabel("Y-Coordinate")
plt.title("Aggregation Dataset")
```

```python
dbscan = DBSCAN(eps=0.25, min_samples=20)
dbscan.fit(data_scaled)
```

```python
labels=dbscan.labels_
print(labels)
```

```python
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=labels,cmap="inferno")
```

```python
silhouette_avg = silhouette_score(data_scaled, dbscan.labels_)

print("Silhouette Index:", silhouette_avg)
```

# SPIRAL DATASET

```python
data_spiral=pd.read_csv("spiral.txt",delimiter="\t",header=None)
data_spiral.columns=["X-Coordinate","Y-Coordinate","Clusters"]
```

```python
data_spiral.drop("Clusters",axis=1,inplace=True)
data_spiral.head()
```

```python
data_spiral_scaled=scaler.fit_transform(data_spiral)
```

```python
plt.scatter(data_spiral_scaled[:,0],data_spiral_scaled[:,1])
plt.xlabel("X-Coordi")
plt.ylabel("Y-Coordi")
plt.title("Spiral Dataset")
```

```python
dbscan_spiral=DBSCAN(eps=0.63,min_samples=35)
dbscan_spiral.fit(data_spiral_scaled)
```

```python
labels_spiral=dbscan_spiral.labels_
print(labels_spiral)
```

```python
plt.scatter(data_spiral_scaled[:, 0], data_spiral_scaled[:, 1], c=labels_spiral,cmap="inferno")
```

```python
silhouette_spiral_avg = silhouette_score(data_spiral_scaled, labels_spiral)

print("Silhouette Index:", silhouette_spiral_avg)
```

# COMPOUND DATASET


```python
data_compound=pd.read_csv("compound.txt",delimiter=",",header=None)
data_compound.columns=["X-Coordinate","Y-Coordinate","Clusters"]
```

```python
data_compound.drop("Clusters",axis=1,inplace=True)
data_compound.head()
```

```python
data_compound_scaled=scaler.fit_transform(data_compound)
```

```python
plt.scatter(data_compound_scaled[:,0],data_compound_scaled[:,1])
plt.xlabel("X-Coordi")
plt.ylabel("Y-Coordi")
plt.title("Compound Dataset")
```

```python
dbscan_com=DBSCAN(eps=0.72,min_samples=40)
dbscan_com.fit(data_compound_scaled)
```

```python
labels_com=dbscan_com.labels_
print(labels_com)
```

```python
plt.scatter(data_compound_scaled[:, 0], data_compound_scaled[:, 1], c=labels_com,cmap="plasma")
```

```python
silhouette_compound_avg = silhouette_score(data_compound_scaled, labels_com)

print("Silhouette Index:", silhouette_compound_avg)
```

```python
sil_categories=["Aggregation Si","Spiral Si","Compound Si"]
sil_scores=[silhouette_avg,silhouette_spiral_avg,silhouette_compound_avg]
plt.bar(sil_categories,sil_scores,color="green")
bars=plt.bar(sil_categories,sil_scores,color="green")
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f'{height:.3f}', ha='center', va='bottom')
plt.xlabel("Datasets")
plt.ylabel("Silhoutte Scores")
plt.title("Silhoutte Score Comparison")
plt.show()
```
